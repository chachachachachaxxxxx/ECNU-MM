## python基础知识
### 三引号
将字符串内容放在三引号里不仅保留字符串内容， 还保留字符串格式，通常多行文字放在三引号中

```python
content="""hello
world!"""
print(content)
```

### 输出print()函数
print(item1[,item2,...,sep=分割字符,end=终止符])
#### 两种格式化输出
print("%sitem%d"%(string, 参数行, int))
print("{}item{}".format(string, 参数行, int))
### 输出input()函数
变量=input(提示字符串)
### range()函数
range([start, ]end[, step])产生一个整数序列
start默认为0, step默认为1

## 复合数据类型
### list[]
#### list元素增加
append: 每次向list尾部添加一个元素
extend: 向list尾部添加多个元素
insert(index, elem): 向list的index位置插入elem
#### list元素删除
pop(index): 省略index即删除末尾元素, 否则删除index位置的元素
remove(elem): 删除指定elem元素, 若有多个, 则删除第一个元素
clear: 将列表清空
