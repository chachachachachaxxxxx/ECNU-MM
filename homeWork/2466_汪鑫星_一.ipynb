```python
import numpy as np
#np.arange([start, ]end[, step], dtype)
a = np.arange(5, dtype=float)
print(a)
b = np.arange(0, 10, 2, dtype=int)
print(b)
#np.empty((rownum, colnum), dtype)
c = np.empty((2, 3), int)
print(c)
#np.linspace(start, end, num of numbers)
d = np.linspace(-1, 2, 5)
print(d)
e = np.random.randint(0,3, (2, 3))
print(e)

#使用虚数j生成数组
f = np.mgrid[0:2:5j] # 等价于np.linspace(0, 2, 5)
x, y = np.mgrid[0:2:4j, 10:20:5j] # 生成[0, 2] * [10, 20]上的4 * 5的数组
print("x={}\ny={}".format(x,y))
```

#### reshape & resize

```python
import numpy as np
a = np.arange(4).reshape(2, 2)
print(a.reshape(4), '\n', a) # reshape不改变自身
# print(a.resize(4), '\n', a) # resize改编自生

# 数组组合
b = np.arange(4, 8).reshape(2, 2)
c1 = np.vstack([a, b]) # 等价于np.r_[a,b]
print(c1)
c2 = np.hstack([a,b]) #等价于np.c_[a,b]
print(c2)

# 数组分割
d1 = np.hsplit(b, 2) # 把b分成两个列数组
d2 = np.vsplit(b, 2) # 把b分成两个行数组

# 四则运算
aplus = a + b
print(aplus)
atimes = a * b
print(atimes)
e1 = np.modf(a / b)[0] # 对应元素相除的小数部分
e2 = np.modf(a / b)[1] # 对应元素相除的整数部分

# np.where
print(np.where(a > 2, -1, a)) # a中大于2的元素改为-1
print(np.where(a > 2, -1, 0)) # a中大于2的元塞努改为-1, 否则0
```

```python
#程序文件Pex2_45.py
import numpy as np
import pandas as pd
from matplotlib.pyplot import *
a=pd.read_excel("Trade.xlsx")
a['year']=a.Date.dt.year  #添加交易年份字段
a['month']=a.Date.dt.month  #添加交易月份字段
rc('font',family='SimHei') #用来正常显示中文标签

ax1=subplot(2,3,1)   #建立第一个子图窗口
Class_Counts=a.Order_Class[a.year==2012].value_counts()
Class_Percent=Class_Counts/Class_Counts.sum()
ax1.set_aspect(aspect='equal')  #设置纵横轴比例相等
ax1.pie(Class_Percent,labels=Class_Percent.index, normalize=False,
        autopct="%.1f%%")  #添加格式化的百分比显示
ax1.set_title("2012年各等级订单比例")

ax2=subplot(232)  #建立第2个子图窗口
#统计2012年每月销售额
Month_Sales=a[a.year==2012].groupby(by='month').aggregate({'Sales':np.sum})
#下面使用Pandas画图
Month_Sales.plot(title="2012年各月销售趋势",ax=ax2, legend=False)
ax2.set_xlabel('')

ax3=subplot(2,3,(3,6))
cost=a['Trans_Cost'].groupby(a['Transport'])
ts = list(cost.groups.keys())
# print(list(map(cost.get_group, ts)))
dd = np.array(list(map(cost.get_group, ts)))
boxplot(dd); gca().set_xticklabels(ts)

ax4=subplot(2,3,(4,5))
hist(a.Sales[a.year==2012],bins=40, density=True)
ax4.set_title("2012年销售额分布图");
ax4.set_xlabel("销售额");
savefig("figure2_45.png"); show()
```

## scipy.stats
包含了多种概率分布的随机变量，随机变量分为连续型和离散型两种

```python
# 绘制4个不同正态分布的密度函数画在一个窗口里
from matplotlib.pyplot import plot, legend, xlabel, ylabel, savefig, show, rc
from scipy.stats import norm
from numpy import linspace
x=linspace(-7,7,100); rc('font',size=15); rc('text', usetex=True)
mu0=[-1,0];s0=[0.5,1]
plot(x,norm(mu0[0], s0[0]).pdf(x),'r*-',label="$\\mu=-1, \\sigma=0.5$")
plot(x,norm(mu0[0], s0[1]).pdf(x),'bp-',label="$\\mu=-1, \\sigma=1$")
plot(x,norm(mu0[1], s0[0]).pdf(x),'k-',label="$\\mu=0, \\sigma=0.5$")
plot(x,norm(mu0[1], s0[1]).pdf(x),'>g-',label="$\\mu=0, \\sigma=1$")
legend(); xlabel('$x$'); ylabel('$f(x)$')
savefig("figure2_46.png",dpi=500); show()
```

### 符号运算
#### evalf() & n()用于获取任何对象的浮点近似值
#### together() & apart()用于实现有理数的加法和除法

```python
#程序文件Pex3_1.py
from sympy import *
x,y,z=symbols('x  y  z')
m0,m1,m2,m3=symbols('m0:4')  #创建多个符号变量
x=sin(1)
print("x=",x); print("x=",x.evalf())
print("x=",x.n(16))  #显示小数点后16位数字
print("pi的两种显示格式:{},{}".format(pi,pi.evalf(3)))  #这里不能使用n()函数
expr1=y*sin(y**2)  #创建第一个符号表达式
expr2=y**2+sin(y)*cos(y)+sin(z)  #创建第二个符号表达式
print("expr1=",expr1)
print("y=5时，expr1=",expr1.subs(y,5))  #代入一个符号变量的值
print("y=2,z=3时，expr2=",expr2.subs({y:2,z:3}))  #代入y=2,z=3
print("y=2,z=3时，expr2=",expr2.subs({y:2,z:3}).n())  #以浮点数显示计算结果
```
